Load Dataset

```python
import pandas as pd
# Load dataset
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)
# Display first few rows
print(df.head())
```

Obtain number of rows and columns

```python
rows, cols = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
```

Get a concise summary of the dataset

```python
print(df.info())
```

Generate descriptive statistics for numerical columns

```python
print(df.describe())
```

Data Exploration:

Print the number of unique Campaign_ID values

```python
unique_campaigns = df["Campaign_ID"].nunique()
print(f"Number of unique Campaign_ID values: {unique_campaigns}")
```

List the unique values of Location and Customer_Segment columns

```python
unique_locations = df["Location"].unique()
unique_segments = df["Customer_Segment"].unique()
print("Unique Location values:", unique_locations)
print("Unique Customer_Segment values:", unique_segments)
```

Count occurrences of each category in Campaign_Type and Channel_Used

```python
campaign_type_counts = df["Campaign_Type"].value_counts()
channel_used_counts = df["Channel_Used"].value_counts()
print("Campaign Type Counts:\n", campaign_type_counts)
print("Channel Used Counts:\n", channel_used_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Acquisition_Cost"], y=df["ROI"], alpha=0.7, edgecolor='k')
# Add labels and title
plt.xlabel("Acquisition Cost")
plt.ylabel("Return on Investment (ROI)")
plt.title("Scatter Plot of Acquisition Cost vs ROI")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)
# Convert 'Date' column to datetime format (handling DD-MM-YYYY format)
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors='coerce')
# Drop rows where Date conversion failed (if any)
df = df.dropna(subset=["Date"])
# Group by Date and Company, then calculate the average Conversion Rate
conversion_trend = df.groupby(["Date", "Company"])["Conversion_Rate"].mean().reset_index()
# Line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Conversion_Rate", hue="Company", data=conversion_trend, marker="o", palette="tab10")
# Add labels and title
plt.xlabel("Date")
plt.ylabel("Average Conversion Rate")
plt.title("Trend of Conversion Rate Over Time for Each Company")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True)
plt.legend(title="Company")
# Show plot
plt.show()
```
